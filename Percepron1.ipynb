```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=500, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
# replace 0s with -1 so that it can contribute in loss calculation
y[y==0] = -1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
scatter = plt.scatter(X[:,0],X[:,1],c=y,cmap='brg',s=10)
plt.legend(handles=scatter.legend_elements()[0], labels=['Class 0', 'Class 1'], title="Target")
plt.show()
```

```python
def perceptron(X,y):

    w1=w2=b=1
    lr = 0.1

    for j in range(1000):

        for i in range(X.shape[0]):

            # check condition
            z = w1*X[i][0] + w2*X[i][1] + b

            if z*y[i] < 0:
                w1 = w1 + lr*y[i]*X[i][0]
                w2 = w2 + lr*y[i]*X[i][1]
                b = b + lr*y[i]

    return w1,w2,b
```

```python
w1,w2,b = perceptron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
b
```

```python
x_input = np.linspace(X[:,0].min()-0.2,X[:,0].max()+0.2,100)
y_input = m*x_input + c

import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
# 1. Add a label to the plot function
line, = plt.plot(x_input, y_input, color='red', linewidth=2, label='Decision Boundary')

scatter = plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=20)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron')
plt.tight_layout()

# 2. Combine the line handle with the scatter handles, and combine their labels
handles = [line] + scatter.legend_elements()[0]
labels = ['Decision Boundary', 'Class 0', 'Class 1']

plt.legend(handles=handles, labels=labels)
plt.show()
```
